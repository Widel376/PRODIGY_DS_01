```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


df = pd.read_csv('mydata.csv',  encoding='utf-8')
df.head()

df_final = df.drop(['Country Code','Country Code','Indicator Name'], axis=1)
country = 'South Africa'


country_data = df_final[df_final['Country Name'] == country]


years = country_data.columns[2:].astype(int)
population = country_data.iloc[:, 2:].values.flatten()

# Bar graph
plt.figure(figsize=(8, 4))
plt.bar(years, population, color='red', width =0.5)
plt.xlabel('Year')
plt.ylabel('Population in (millions)')
plt.title(f'Population of females in {country} over the Years')


plt.show()

# Histogram
plt.figure(figsize=(8, 5))
plt.hist(population, bins=20, color='green', edgecolor='black')
plt.xlabel('Population in (millions)')
plt.ylabel('Frequency')
plt.title(f'Distribution of Population of females in {country}')
plt.show()
     
```
